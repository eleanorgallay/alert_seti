```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
#defining path to folder with my notebook and gaia data
path = '/Users/eleanorgallay/Google Drive/SETIsummer/' 
```

```python
#reading csv file
df = pd.read_csv(path + 'gaia.tsv',comment="#",delimiter=";")
```

```python
#dropping empty rows and rows with units rather than numbers
gaia = df.drop([0,1])
```

```python
#defining ra and dec coordinates from our gaia dataframe
ra = pd.to_numeric(gaia['RA_ICRS'])
dec = pd.to_numeric(gaia['DE_ICRS'])
```

```python
#making a scatterplot!
plt.figure(figsize=(10, 8))
plt.plot(ra, dec)
plt.xlabel('Right Ascension (degrees)')
plt.ylabel('Declination (degrees)')
plt.title('Positions of Gaia Stars (Scatterplot)')
plt.show()
print('oh no this looks terrible!')
```

```python
#making a 2D Histogram!
plt.figure(figsize=(10, 8))
plt.hist2d(ra, dec, bins=(200, 50), cmap=plt.cm.coolwarm)
plt.xlabel('Right Ascension (degrees)')
plt.ylabel('Declination (degrees)')
plt.title('Positions of Gaia Stars (2D Histogram)')
plt.colorbar()
plt.show()
print("That's more like it")
```

```python
#making a CMD!

BP = pd.to_numeric(gaia['BPmag'], errors='coerce') #BP magnitude
RP = pd.to_numeric(gaia['RPmag'], errors='coerce') #RP magnitude
color = BP - RP #BP-RP color
gmag = pd.to_numeric(gaia['Gmag'], errors='coerce') #apparent g magnitudes
plx = pd.to_numeric(gaia['Plx'], errors= 'coerce')/1000 #parallax in arcsec
d = 1/plx #distance in parsecs
GMAG = gmag - 5*(np.log10(d/10)) #absolute g magnitudes using distance modulus and parallax
```

```python
#now for plotting...
plt.figure(figsize=(10, 8))
plt.plot(color, GMAG, '.', markersize=0.5)
plt.xlabel('BP-RP color')
plt.ylabel('Absolute G Magnitude')
plt.xlim(-1, 5.5)
plt.ylim(20, -2)
plt.title('Gaia Stars Color-Magnitude Diagram')
plt.show()
print('All done!')
```
